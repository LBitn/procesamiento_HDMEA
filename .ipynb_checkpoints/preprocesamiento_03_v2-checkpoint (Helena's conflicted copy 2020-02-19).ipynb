```julia
SCPpath = "/home/isabel/Dropbox/Proyects/Romero-Gutierrez/SCP/" # path to SCPs
cd( SCPpath )
push!( LOAD_PATH, "." ); # folder que continen los modulos. Correr desde ahí dentro.
using AllFunctions # modulo personal
#
using JLD, DSP;
```

```julia
BRWname = "/home/isabel/Desktop/200819/acido_01.brw"
SamplingRate = brw_things( BRWname )[ "SamplingRate" ];
Γ = OS( "linux" );
JLDfolder = "/home/isabel/Desktop/200819/acido_01/NoSat/";
JLDfiles = find_files( JLDfolder, "jld", Γ);
cte = 3;
responsetype = Bandstop( 50, 70; fs = SamplingRate );
designmethod = Butterworth( 5 );
```

```julia
CHindex = size( ( brw_things( BRWname )[ "Chs" ] ), 2 );
CHnumber = size( ( brw_things( BRWname )[ "Chs" ] ), 1 );
ALLvaluesRAW = zeros( Int, CHnumber, ( size( JLDfiles, 1 ) + CHindex + 1 ) );
ALLvaluesRAW[ :, 1:CHindex ] = brw_things( BRWname )[ "Chs" ];
ALLvaluesFILT = zeros( Int, CHnumber, ( size( JLDfiles, 1 ) + CHindex + 1 ) );
ALLvaluesFILT[ :, 1:CHindex ] = brw_things( BRWname )[ "Chs" ];
```

```julia
using Statistics
function find_events( data, n )
    allEvents = Dict( );
    thr = n*( -1*median( abs.( data ) )/0.6745 );
    Events = findall( x -> x <= thr, data );
    channels = zeros( Int, size( Events, 1 ) );
    Frames = zeros( Int, size( Events, 1 ) );
    for i = 1:size( Events, 1 )
        channels[ i ] = Int( Events[ i ][ 1 ] );
        Frames[ i ] = Int( Events[ i ][ 2 ] );
    end
    order_channs = sort( unique( channels ) );
    for i = 1:size( order_channs, 1 )
        order_Frames = Frames[ order_channs[ i ] .== channels ];
        temp = Dict(
            "channel" => lpad( string( order_channs[ i ]), 4, "0" ),
            "N_events" => size( order_Frames, 1 ),
            "Frames" => order_Frames
            );
        get!( allEvents, i, temp )
    end
    return allEvents
end
```

```julia
for I = 1:size( JLDfiles, 1 );
    JLDname = JLDfiles[ I ];
    data =  load( JLDname ); # carga el bin i
    RAWdata = data[ "data" ];
    FILTdata = filt( digitalfilter( responsetype, designmethod ), RAWdata );
    RAWevents = find_events( RAWdata, cte );
    FILTevents = find_events( FILTdata, cte );
    for i = 1:length( RAWevents )
        AE = RAWevents[ i ];
        ALLvaluesRAW[ 
            ALLvaluesRAW[ :, 1 ] .== parse( Int, AE[ "channel" ] ), CHindex + I 
            ] .=  AE[ "N_events" ];
    end
    for i = 1:length( FILTevents )
        AE = FILTevents[ i ];
        ALLvaluesFILT[ 
            ALLvaluesFILT[ :, 1 ] .== parse( Int, AE[ "channel" ] ), CHindex + I 
            ] .=  AE[ "N_events" ];
    end
end
```

```julia
ALLvaluesFILT[:,end] = sum(ALLvaluesFILT[:,4:end-1],dims = 2);
```
